### Starting the Tutorial ( finding out the flower)

```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
iris.data[0:5]
```

#### Making a dataframe using the data and the feature names

```python
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df.head()
```

```python
df['target'] = iris.target
df.head()
```

0 means setosa, 1 means versicolor and 2 means virginica  

```python
iris.target_names 
```

### Finding rows which has the target value 1

```python
df[df.target==1].head()
```

```python
df[df.target==2].head()
```

```python
df[df.target==0].head()
```

### Adding the flower name  based on the values 0,1 and 2

```python
df['flower_name'] = df.target.apply(lambda x: iris.target_names[x])
df.head()
```

```python
from matplotlib import pyplot as plt
```

```python
%matplotlib inline
```

### Seperating the three target flowers

```python
df0 = df[df.target==0]
df1 = df[df.target==1]
df2 = df[df.target==2]
```

#### Drawing the Scatter plot

### Training the model

```python
from sklearn.model_selection import train_test_split
```

#### Dropping the target and flower name from the input
#### only the features should be in the input dataframe

```python
X = df.drop(['target','flower_name'], axis ='columns' ) 
X.head()
```

#### Only the numerical target values in the y 

```python
y = df.target
y.head()
```

#### Spliting dataset

```python
X_train, X_test, y_train, y_test =train_test_split(X, y, test_size= 0.2) # X_train, X_test = X 
```

#### Importing and fit

```python
from sklearn.svm import SVC
model = SVC() # Shift+tab to see the infos
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test,y_test) # the model will predict the value based on X_test
                            # and compare the value will actual y_test
```
