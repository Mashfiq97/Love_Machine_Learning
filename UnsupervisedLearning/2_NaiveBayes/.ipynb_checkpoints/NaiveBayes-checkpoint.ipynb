### Starting From Here (Finding the survival rate)

```python
import pandas as pd
df= pd.read_csv("titanic.csv")
df.head()
```

### Droping the columns which are not relevent with the survival rate

```python
df.drop(['PassengerId', 'Name', 'SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked'], axis= 'columns', inplace=True) # updating the dataframe
df.head()
```

#### Separating the Target Column and Input Column

```python
target = df.Survived # saving
inputs = df.drop('Survived', axis = 'columns') # removing
inputs
```

#### Creating Dummy Variables for the Sex Column

```python
dummies = pd.get_dummies(inputs.Sex)
dummies.head()
```

#### Appending these columns in the Inputs 

```python
inputs = pd.concat([inputs, dummies],axis='columns')
inputs.head()
```

```python
inputs.drop('Sex', axis = 'columns', inplace=True) # removing the Sex Column
inputs.head(3)
```

### Finding any NaN values in a Column

```python
inputs.columns[inputs.isna().any()] # "Age" column has NaN values
```

```python
inputs.Age[:10]
```

#### Filling the NaN values with the mean

```python
inputs.Age = inputs.Age.fillna(inputs.Age.mean())
inputs.head(10)
```

#### Spliting the dataset

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs, target, test_size=0.2) 
```

```python
len(inputs)
```

```python
len(X_train)
```

```python
len(X_test)
```

### Creating the Gaussian Model as the dataset distribution is normal

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
model.predict(X_test[:10])
```

```python
model.predict_proba(X_test[:10]) # first column indicates "Did not survive"
```

## Second Part: Spam Detection

```python
import pandas as pd
df = pd.read_csv("spam.csv")
df.head()
```

#### Grouping based on Category

```python
df.groupby('Category').describe()
```

#### Adding a new column named Spam

```python
df['spam']=df['Category'].apply(lambda x: 1 if x=='spam' else 0)
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.Message, df.spam, test_size=0.25) 
```

#### Converting the Message column into numeric values (SKLearn CountVectorizer)

#### Converting Unique words into numbers


```python
from sklearn.feature_extraction.text import CountVectorizer

v = CountVectorizer()

X_train_count = v.fit_transform(X_train.values) # X_train values to X_train_count
X_train_count.toarray()[0:5]
```

#### Training the model

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train_count, y_train)
```

#### Detecting Emails

```python
emails = [
    'Hey mohan, can we get together to watch footbal game tomorrow?',
    'Upto 20% discount on parking, exclusive offer just for you. Dont miss this reward!'
]
emails_count = v.transform(emails)

model.predict(emails_count) # Second Email Spam
```

```python
X_test_count = v.transform(X_test) # convert the words into numbers
model.score(X_test_count, y_test)
```

### sklearn Pipeline

#### Now doing the things again using Pipeline in a simple way
#### Pipeline has been used to do some works sequentially

```python
from sklearn.pipeline import Pipeline

clf = Pipeline([
    ('vectorizer', CountVectorizer()), 
    ('nb', MultinomialNB())
])
```

```python
clf.fit(X_train, y_train)
```

```python
clf.score(X_test,y_test)
```

```python
clf.predict(emails)
```
