```python
#loading the dataset
import pandas as pd
df = pd.read_csv("breast cancer classification dataset.csv")
df.head(10)
```

```python
# Getting the rows and column count in the dataset
df.shape  # patients(rows) = 569 and features(columns)=33
```

```python
# Getting the total number of empty values in each column
df.isna().sum()
```

```python
# Droping the "Unnamed: 32" column
df = df.drop('Unnamed: 32', axis= 'columns')
```

```python
df.head(5)
```

```python
# new Shape
df.shape # dropped one column
```

```python
# Handling the Missing values
import math 
median_radius = math.floor(df.radius_mean.median())
#print(median_radius)
```

```python
#inserting values to NaN of radius_mean
df.radius_mean = df.radius_mean.fillna(median_radius) 
df.head(5)
```

```python
median_fractal_dimension_worst = math.floor(df.fractal_dimension_worst.median())
#print(median_fractal_dimension_worst)
```

```python
##inserting values to NaN of fractal_dimension_worst column
df.fractal_dimension_worst = df.fractal_dimension_worst.fillna(median_fractal_dimension_worst)
```

```python
df.isna().sum() # no NaN values 
```

```python
#Using One Hot Encoding in the diagnosis column
dummies= pd.get_dummies(df.diagnosis)
dummies.head(5)
```

```python
# concatanating these columns in the original dataframe
merged = pd.concat([df, dummies], axis='columns')  
merged.head(20)
```

```python
# droping the diagnosis and B columns
final_data_frame = merged.drop(['diagnosis', 'B'], axis='columns')
final_data_frame.head(20)
```

```python
#Dropping the id column
final_data_frame = final_data_frame.drop('id', axis= 'columns') 
final_data_frame.head(20)
```

```python
final_data_frame.tail(10)
```

```python
#Split the dataset into features and labels
X = final_data_frame.drop('M', axis= 'columns')
y = final_data_frame['M']
```

```python
X.head(5)
```

```python
y.head(5)
```

```python
#Scaling the Input Columns only
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler()
X_scaled = min_max_scaler.fit_transform(X)
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)
X_scaled_df.head(5)
```

```python
X = X_scaled_df.copy()
X.head(5)

```

```python
y.head(5)
```

### Splitting the Dataset

```python
#Spilitng the dataset to train test dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y ,test_size=0.2, random_state=1)
```

```python
X.shape
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

### Importing the models

```python
from sklearn.svm import SVC
model_SVC = SVC(kernel = 'linear')
model_SVC.fit(X_train, y_train)
```

```python
from sklearn.neural_network import MLPClassifier
nnc = MLPClassifier(hidden_layer_sizes=(7), activation="relu", max_iter=10000)
nnc.fit(X_train, y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
model_RFC = RandomForestClassifier(n_estimators = 50, criterion = 'entropy')
model_RFC.fit(X_train, y_train)
```

### Performing PCA

```python
X.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
X_scaled_new = scaler.fit_transform(X)
```

```python
from sklearn.decomposition import PCA 
pca = PCA(n_components=15)
principal_components = pca.fit(X_scaled_new)
```

```python
print(principal_components)
```

```python
sum(pca.explained_variance_ratio_)
```

```python
transformed_df = pd.DataFrame(pca.transform(X_scaled_new), 
                              columns=['PC1', 'PC2', 'PC3', 'PC4', 'PC5', 
                                       'PC6', 'PC7', 'PC8', 'PC9', 'PC10',
                                       'PC11', 'PC12', 'PC13', 'PC14', 'PC15'])

transformed_df.head(5)
```

```python
# transformed_df = X
X_train_PCA, X_test_PCA, y_train_PCA, y_test_PCA = train_test_split(transformed_df, y ,test_size=0.2, random_state=1)
```

```python
X_train_PCA.shape
```

```python
X_test_PCA.shape
```

```python
transformed_df.shape
```

```python
model_SVC_PCA = SVC(kernel = 'linear')
model_SVC_PCA.fit(X_train_PCA, y_train_PCA)
```

```python
nnc_PCA = MLPClassifier(hidden_layer_sizes=(7), activation="relu", max_iter=10000)
nnc_PCA.fit(X_train_PCA, y_train_PCA)
```

```python
model_RFC_PCA = RandomForestClassifier(n_estimators = 50, criterion = 'entropy')
model_RFC_PCA.fit(X_train_PCA, y_train_PCA)
```

### Comparing the accuracies of Models

```python
from sklearn.metrics import accuracy_score
y_pred = model_SVC.predict(X_test) 
acc_SVC = accuracy_score(y_pred, y_test)
acc_SVC
```

```python
y_pred_PCA = model_SVC_PCA.predict(X_test_PCA) 
acc_SVC_PCA = accuracy_score(y_pred_PCA, y_test_PCA)
acc_SVC_PCA
```

```python
#Ploting the accuracy
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
mdl = ['Pre PCA', 'Post PCA']
acc = [acc_SVC, acc_SVC_PCA]
ax.bar(mdl,acc)
plt.ylabel("Accuracy")
plt.title("SVM Accuracy on Breast Cancer Dataset")
plt.show()
```

```python
y_pred = nnc.predict(X_test) 
acc_nnc = accuracy_score(y_pred, y_test)
acc_nnc
```

```python
y_pred_PCA = nnc_PCA.predict(X_test_PCA) 
acc_nnc_PCA = accuracy_score(y_pred_PCA, y_test_PCA)
acc_nnc_PCA
```

```python
#Ploting the accuracy
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
mdl = ['Pre PCA', 'Post PCA']
acc = [acc_nnc, acc_nnc_PCA]
ax.bar(mdl,acc)
plt.ylabel("Accuracy")
plt.title("Multilayer Perceptron Classifier Accuracy on Breast Cancer Dataset")
plt.show()
```

```python
y_pred = model_RFC.predict(X_test) 
acc_model_RFC = accuracy_score(y_pred, y_test)
acc_model_RFC
```

```python
y_pred_PCA = model_RFC_PCA.predict(X_test_PCA) 
acc_model_RFC_PCA = accuracy_score(y_pred_PCA, y_test_PCA)
acc_model_RFC_PCA
```

```python
#Ploting the accuracy
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
mdl = ['Pre PCA', 'Post PCA']
acc = [acc_model_RFC, acc_model_RFC_PCA]
ax.bar(mdl,acc)
plt.ylabel("Accuracy")
plt.title("RFC Accuracy on Breast Cancer Dataset")
plt.show()
```
