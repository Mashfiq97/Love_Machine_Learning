### Predicting the Price

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv("MultiValue.csv")
df
```

### Handling the Missing value of Bedroom Table

```python
import math 
median_bedrooms = math.floor(df.bedrooms.median())
print(median_bedrooms)
```

### Inserting the missing value

```python
df.bedrooms = df.bedrooms.fillna(median_bedrooms) #updating the dataframes
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area','bedrooms','age']], df.price) #Independent_Variable, Target_Variable
```

### Model is Ready

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[3000,3,40]]) #size,bedrooms,age
```
