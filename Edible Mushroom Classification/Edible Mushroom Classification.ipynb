```python
#loading the dataset
import pandas as pd
df = pd.read_csv("mushroom edibility classification dataset.csv")
df.head(10)
```

```python
# Getting the rows and column count in the dataset
df.shape
```

```python
# Getting the total number of empty values in each column
df.isna().sum()
```

```python
# Dropping the "Unnamed: 32" column
df = df.drop('Unnamed: 0', axis= 'columns')
```

```python
df.head(5)
```

```python
# new Shape
df.shape # dropped one column
```

```python
# Handling the Missing values
import math 
cap_shape_median = math.floor(df["cap-shape"].median())
print(cap_shape_median)
```

```python
df["cap-shape"] = df["cap-shape"].fillna(cap_shape_median) 
df.head(5)
```

```python
cap_color_median = math.floor(df["cap-color"].median())
print(cap_color_median)
```

```python
df["cap-color"] = df["cap-color"].fillna(cap_color_median ) 
df.head(5)
```

```python
# Getting the total number of empty values in each column
df.isna().sum()
```

```python
# Converting Categorical Value to Numerical Value
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
for column in df.columns:
    df[column] = labelencoder.fit_transform(df[column])
```

```python
df.head(10)
```

### Dropping the Columns which has Null Values

```python
df = df.drop('veil-type', axis= 'columns')
```

```python
df.tail()
```

```python
df = df.drop('veil-color', axis= 'columns')
```

```python
df = df.drop('ring-number', axis= 'columns')
```

```python
df.head(5)
```

### Split the dataset into features and labels

```python
X = df.drop(["class"], axis= 'columns')
X.head(5)
```

```python
Y = df['class']
Y.head(5)
```

### Scaling the Column Values

```python
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler()
X_scaled = min_max_scaler.fit_transform(X)
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)
X_scaled_df.head(5)
```

```python
X = X_scaled_df.copy()
X.head(5)
```

```python
#Splitting the dataset to train test dataset
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y ,test_size=0.2, random_state=1)
```

```python
X.shape
```

```python
X_train.shape
```

```python
X_test.shape
```

### Importing the models

```python
from sklearn.svm import SVC
model_SVC = SVC(kernel = 'linear')
model_SVC.fit(X_train, Y_train)
```

```python
from sklearn.neural_network import MLPClassifier
nnc = MLPClassifier(hidden_layer_sizes=(7), activation="relu", max_iter=10000)
nnc.fit(X_train, Y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
model_RFC = RandomForestClassifier(n_estimators = 50, criterion = 'entropy')
model_RFC.fit(X_train, Y_train)
```

### Performing PCA

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
X_scaled_new = scaler.fit_transform(X)
```

```python
from sklearn.decomposition import PCA 
pca = PCA(n_components=10)
principal_components = pca.fit(X_scaled_new)
```

```python
print(principal_components)
```

```python
transformed_df = pd.DataFrame(pca.transform(X_scaled_new), 
                              columns=['PC1', 'PC2', 'PC3', 'PC4', 'PC5', 
                                       'PC6', 'PC7', 'PC8', 'PC9', 'PC10'])

transformed_df.head(5)
```

```python
X_train_PCA, X_test_PCA, y_train_PCA, y_test_PCA = train_test_split(transformed_df, Y ,test_size=0.2, random_state=1)
```

```python
X_train_PCA.shape
```

```python
transformed_df.shape
```

### After applying PCA procedure, now training the models


```python
model_SVC_PCA = SVC(kernel = 'linear')
model_SVC_PCA.fit(X_train_PCA, y_train_PCA)
```

```python
nnc_PCA = MLPClassifier(hidden_layer_sizes=(7), activation="relu", max_iter=10000)
nnc_PCA.fit(X_train_PCA, y_train_PCA)
```

```python
model_RFC_PCA = RandomForestClassifier(n_estimators = 50, criterion = 'entropy')
model_RFC_PCA.fit(X_train_PCA, y_train_PCA)
```

### Comparing the accuracies of pre PCA and post PCA Models

```python
from sklearn.metrics import accuracy_score
Y_pred = model_SVC.predict(X_test) 
acc_SVC = accuracy_score(Y_pred, Y_test)
acc_SVC
```

```python
y_pred_PCA = model_SVC_PCA.predict(X_test_PCA) 
acc_SVC_PCA = accuracy_score(y_pred_PCA, y_test_PCA)
acc_SVC_PCA
```

### Plotting the accuracy

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
mdl = ['Pre PCA', 'Post PCA']
acc = [acc_SVC, acc_SVC_PCA]
ax.bar(mdl,acc)
plt.ylabel("Accuracy")
plt.title("SVM Accuracy on Edible Mushroom Classification Dataset")
plt.show()
```
